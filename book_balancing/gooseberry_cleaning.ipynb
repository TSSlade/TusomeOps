```python
import pandas as pd
```

```python
df = pd.read_excel("gooseberry_records_2018-10-17.xlsx",
                   usecols="A:P")

df.rename(columns={"Reason for deficit : New school?, increased enrollment ?" : "reason",
                   "Confirmation of how its was resolved": "resolution",
                  "Region": "region",
                  "teachers_guides": "rw_tgs",
                  "pupils_books": "rw_pbs",
                  "enrollment": "rw_enroll"}, inplace=True)
```

```python
df.resolution.unique(), df.reason.unique()
```

```python
df.drop(["resolution","reason"], axis = 1, inplace=True)
```

```python
df.rw_enroll.head(20)
```

```python
df["g1_parens"] = df.rw_enroll.str.extract("(1\s*\(.*\))", expand=True)
df["g2_parens"] = df.rw_enroll.str.extract("(2\s*\(.*\))", expand=True)
df["g3_parens"] = df.rw_enroll.str.extract("(3\s*\(.*\))", expand=True)

df[["rw_enroll","g1_parens","g2_parens","g3_parens"]]
```
